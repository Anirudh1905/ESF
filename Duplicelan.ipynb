```
import pandas as pd
```

```
df=pd.read_excel('/content/drive/My Drive/Ramnagar.xlsx')
df.head()
```

```
df.fillna('unknown',inplace=True)
df.head()
```

```
len(df)
```

```
q=[]
for i,a,b,c,d in zip(range(len(df)),df['unclean_village'],df['unclean_name'],df['unclean_relative_name'],df['unclean_phone_no']):
  for j,e,f,g,h in zip(range(len(df)),df['unclean_village'],df['unclean_name'],df['unclean_relative_name'],df['unclean_phone_no']):
    if i!=j and b==f and c==g and d==h and b!='unknown' and c!='unknown' and d!='unknown':
      print(i,j,a,e,b,f,c,g,d,h)
      q.append(i)    
```

```
k=df[df[['unclean_name','relative_name','unclean_phone_no']].duplicated() == True].index
k
```

```
df.fillna('unknown',inplace=True)

len(df)
```

```
import re
s=[];c=0
for x in df['unclean_years_of_exp'].values:
  if x.find('M')!=-1 or x.find('рдорд╛')!=-1 or x.find('ekg')!=-1 or x.find('m')!=-1:
    p=re.findall("\d+", x)
    for q in p:
      s.append(round((int(q)/12),2)) 
  else:
    if x=='unknown':
      s.append('u')
    else:
      w=re.findall("\d+", x)
      for z in w:
        s.append(int(z))     
print(s)   
```
